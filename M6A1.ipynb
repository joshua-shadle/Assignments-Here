```python
#15.1
import multiprocessing
import time
import random
from datetime import datetime

def show_time():
    time.sleep(random.random())      # wait 0–1 seconds
    print("Process", multiprocessing.current_process().pid,
          "time:", datetime.now())

if __name__ == "__main__":
    for n in range(3):
        p = multiprocessing.Process(target=show_time)
        p.start()


```
