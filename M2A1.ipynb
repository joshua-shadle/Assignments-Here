```python
#Things to Do 4.2
#Purpose: If the named vegetables are equal to the description ,  then it will print that vegetable's name as the result.

small = True
green = False

if small and green:
    print("pea")
elif small and not green:
    print("cherry")
elif green and not small:
    print("watermelon")
else:
    print("pumpkin")
```

```python
#Question 6.2
#Purpose: A while loop that iterates through until it finds the matching number
guess_me = 7
number = 1
while True:
    if number == guess_me:
        print("found it")
        break
    elif number < guess_me:
        print("too low")
    else:
        print("oops")
        break
    number += 1
```

```python
#Question 6.3
#Purpose: A for loop that iterates through a range of numbers to finds the number.
guess_me = 5
for number in range(10):
    if number < guess_me:
        print("too low")
        continue
    elif number == guess_me:
        print("found it")
        break
    else:
        print("oops")
```
